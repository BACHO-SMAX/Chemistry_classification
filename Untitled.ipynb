```python
import pandas as pd
import requests

file_path = r'O:\chemistry\Душистые в-ва исх.xlsx'
df_fragrances = pd.read_excel(file_path)

unique_descriptions = df_fragrances['Описание 350 крт'].dropna().drop_duplicates()
sample_size = min(500, len(unique_descriptions))
sample_texts = unique_descriptions.sample(sample_size, random_state=42).tolist()

batch_size = 30
batches = [sample_texts[i:i + batch_size] for i in range(0, len(sample_texts), batch_size)]

# Убираем идеальный шаблон. Пусть начинает с нуля и потеет!
current_categories = "Список пуст. Начни создавать его с первой партии."

def update_categories(text_batch, current_list, batch_num):
    url = "http://127.0.0.1:1234/v1/chat/completions"
    
    PROMPT = f"""Ты — химический аналитик-детализатор. Твоя цель — создать максимально ПОДРОБНЫЙ классификатор душистых веществ (нужно собрать 25-30 узких бизнес-категорий).
    
    ТЕКУЩИЙ НАКОПЛЕННЫЙ СПИСОК:
    {current_list}
    
    НОВЫЕ ТОВАРЫ ДЛЯ АНАЛИЗА (Партия {batch_num}):
    {text_batch}
    
    ЖЕСТКИЙ ПРИКАЗ:
    1. Изучи партию. Найди в ней специфичные ниши (например: ароматизаторы для табака/вейпов, отдушки для автохимии, коптильные смеси, добавки для кормов, сырье для напитков, чистые химические изоляты и т.д.).
    2. Ты ОБЯЗАН добавить как минимум 1-2 новые УЗКИЕ категории к текущему списку на основе новых товаров. 
    3. НЕ УКРУПНЯЙ! Если видишь пищевые — раздели их на "для выпечки", "для напитков", "для соусов".
    4. Выведи ВЕСЬ обновленный список целиком в виде маркерного списка: "- [Узкая категория] : [Кратко, что сюда входит]". Никакого лишнего текста!
    """

    payload = {
        "model": "qwen2.5-7b-instruct",
        "messages": [
            {"role": "user", "content": PROMPT}
        ],
        "temperature": 0.5, # Оставляем креативность для придумывания новых названий рубрик
        "max_tokens": 1500  # Даем много места для длинного списка
    }
    
    try:
        response = requests.post(url, json=payload)
        if response.status_code != 200:
            return f"Ошибка сервера: {response.text}"
            
        data = response.json()
        if 'choices' in data:
            return data['choices'][0]['message']['content'].strip()
        else:
            return "Ошибка формата ответа."
    except Exception as e:
        return f"Ошибка связи: {e}"

print(f"Подготовлено {len(sample_texts)} строк. Разбито на {len(batches)} партий.")
print("Запускаю форсированный сбор узких категорий...\n")

for i, batch in enumerate(batches):
    print(f"Пытаю нейросеть партией {i+1} из {len(batches)}...")
    batch_text = "\n".join([f"{j+1}. {text}" for j, text in enumerate(batch)])
    
    result = update_categories(batch_text, current_categories, i+1)
    
    if "Ошибка" in result:
        print(f"❌ ОСТАНОВКА: {result}")
        break
    else:
        current_categories = result
        print(f"✅ Партия {i+1} обработана! Список растет.")

print("\n==================================================")
print("ФИНАЛЬНЫЙ РАСШИРЕННЫЙ КЛАССИФИКАТОР (ПОСЛЕ 500 СТРОК):")
print("==================================================")
print(current_categories)
```

```python
import pandas as pd
import requests
import json
import re

# 1. Загружаем файл
file_path = r'O:\chemistry\Душистые в-ва исх.xlsx'
df = pd.read_excel(file_path)

# ДЛЯ ТЕСТА берем 20 строк. Для финала замените на: df_work = df.copy()
df_work = df.head(20).copy()

column_to_analyze = 'Описание 350 крт'

# 2. ФИНАЛЬНЫЙ ПРОМПТ (Зашиваем наш список 25 категорий)
SYSTEM_PROMPT = """Ты — эксперт-сортировщик химических товаров. Твоя задача — классифицировать описания по списку.
Верни строго JSON с 1 полем: "Продукт единый".

СПИСОК КАТЕГОРИЙ (выбирай СТРОГО из него):
1. Ароматизаторы пищевые (жидкие)
2. Ароматизаторы пищевые (сухие/порошок)
3. Ароматизаторы спиртосодержащие (напитки)
4. Ароматизаторы коптильные (Жидкий дым)
5. Ароматизаторы для табачной промышленности
6. Ароматизаторы для вейпов и ЭСДН
7. Ароматические капсулы/сферики (табак)
8. Отдушки для парфюмерии (базы/композиции)
9. Отдушки для косметики и мыла
10. Отдушки для бытовой химии
11. Отдушки для автохимии
12. Эфирные масла (натуральные/нативные)
13. Эфирные масла (реконструированные/синтетика)
14. Растительные экстракты и олеорезины
15. Химические изоляты (чистые в-ва: ментол, ванилин)
16. Смеси душистых веществ (пром. сырье)
17. Препараты для кондитерских изделий и жвачки
18. Препараты для производства соусов и кетчупов
19. Сырье и концентраты для напитков
20. Вспомогательные в-ва (растворители/носители)
21. Пищевые добавки комплексные
22. Кормовые добавки (животные)
23. Ароматизаторы фармацевтические
24. Образцы для лабораторных исследований
25. Прочее (не классифицировано)

ЛОГИКА:
- Капсулы/сферики 3.5мм -> "Ароматические капсулы/сферики (табак)"
- Дым, коптильный -> "Ароматизаторы коптильные (Жидкий дым)"
- Для сигарет, спирт <9% -> "Ароматизаторы для табачной промышленности"
- Lavandula, нативное, из соцветий -> "Эфирные масла (натуральные/нативные)"
"""

def classify_row(text):
    if pd.isna(text) or text == "": return "Пусто"
    url = "http://127.0.0.1:1234/v1/chat/completions"
    payload = {
        "model": "qwen2.5-7b-instruct",
        "messages": [
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": f"Вход: {text}"}
        ],
        "temperature": 0.0 # Робот-сортировщик
    }
    try:
        response = requests.post(url, json=payload)
        res = response.json()['choices'][0]['message']['content']
        match = re.search(r'\{.*\}', res, re.DOTALL)
        parsed = json.loads(match.group(0)) if match else json.loads(res)
        return parsed.get("Продукт единый", "Прочее (не классифицировано)")
    except: return "Ошибка"

# 3. ЗАПУСК
print("Начинаю массовую разметку...")
df_work['Продукт единый'] = df_work[column_to_analyze].apply(classify_row)

print("Разметка завершена!")
display(df_work[[column_to_analyze, 'Продукт единый']])

# Сохранение:
# df_work.to_excel(r'O:\chemistry\Результат_7000_строк.xlsx', index=False)
```

```python
import pandas as pd
import requests
import json
import re

# 1. Загружаем ваш файл (не забудьте про букву r перед путем)
file_path = r'O:\chemistry\data.xlsx'
df = pd.read_excel(file_path)

# БЕРЕМ ТОЛЬКО ПЕРВЫЕ 10 СТРОК ДЛЯ ТЕСТА
# (когда убедитесь, что всё работает идеально, удалите или закомментируйте эту строку)
df_test = df.head(10).copy()

# Название вашей колонки с исходным текстом
column_to_analyze = 'Метка данных'

# 2. Идеальный Мега-Промпт (Конституция для нейросети)
SYSTEM_PROMPT = """Ты — ведущий химический аналитик. Твоя задача — классифицировать таможенные описания производных целлюлозы. 
Верни строго валидный JSON с 1 полем: "Продукт единый". Никакого лишнего текста.

ПРАВИЛО: Выбери СТРОГО ОДНО значение из списка категорий ниже (копируй символ в символ, буква 'Ц' означает Целлюлоза). 
Если сомневаешься или продукта нет в списке, пиши "Прочее (не эфиры/МКЦ или н.д.)".

РАЗРЕШЕННЫЙ СПИСОК КАТЕГОРИЙ:
- Ацетат фталат Ц
- Ацетат Ц
- Ацетопропинат Ц (АПЦ)
- Бутират ацетат Ц
- Гидроксипропил Ц(ГПЦ) и произв
- Гидроксипропилметил Ц (ГПМЦ) и произв
- Гидроксипропилметил Ц фталат
- Гидроксиэтил Ц(ГЭЦ) и произв
- Готовые смеси для оболочек лекарств
- ГПМЦ + Мод. Крахмал
- Диэтиламиноэтил Ц (DEAE)
- Карбоксиметил Ц(КМЦ) и произв
- Коллойдная Ц
- Метил Ц + крахмал
- Метил Ц и произв 
- Метилгидроксиэтил Ц(МГЭЦ)
- Микрокрист Ц(МКЦ)
- н.д. эфиры Ц 
- Нитро Ц и произв
- Поликватрениумы (поликатион Ц)
- Прочее (не эфиры/МКЦ или н.д.)
- Смесь Микрокрист Ц(МКЦ) и Карбоксиметил Ц(КМЦ)
- Смешаные грузы эфиров
- Триацетат Ц
- Этил Ц
- Этилгидрокси этил Ц

ПОДСКАЗКИ ДЛЯ КЛАССИФИКАЦИИ:
- "Кроскармеллоза", "CMC", "Карбоксиметил", "BLANOSE" -> "Карбоксиметил Ц(КМЦ) и произв"
- "HPMC", "Гипромеллоза", "METOLOSE", "METHOCEL" -> "Гидроксипропилметил Ц (ГПМЦ) и произв"
- "HEC", "Гидроксиэтилцеллюлоза", "NATROSOL", "CELLOSIZE" -> "Гидроксиэтил Ц(ГЭЦ) и произв"
- "Polyquaternium", "ПОЛИКВАТЕРНИУМ", "DEKAQUAT" -> "Поликватрениумы (поликатион Ц)"
- "СУХАЯ СМЕСЬ ДЛЯ ПЛЕНОЧНОГО ПОКРЫТИЯ", "AQUAPOLISH" -> "Готовые смеси для оболочек лекарств"
- "МЕТИЛГИДРОКСИЭТИЛЦЕЛЛЮЛОЗА", "AQUENCE LAM" -> "Метилгидроксиэтил Ц(МГЭЦ)"

ПРИМЕР 1:
Вход: "AC-DI-SOL SD-711 CROSCARMELLOSE SODIUM - КРОСКАРМЕЛЛОЗА НАТРИЯ В ПЕРВИЧНОЙ ФОРМЕ..."
Выход: {"Продукт единый": "Карбоксиметил Ц(КМЦ) и произв"}

ПРИМЕР 2:
Вход: "AQUAPOLISH СУХАЯ СМЕСЬ ДЛЯ ПЛЕНОЧНОГО ПОКРЫТИЯ ОБОЛОЧКОЙ..."
Выход: {"Продукт единый": "Готовые смеси для оболочек лекарств"}
"""

# 3. Функция общения с локальной нейросетью
def ask_qwen(text):
    if pd.isna(text) or text == "":
        return "Пусто"
    
    url = "http://127.0.0.1:1234/v1/chat/completions"
    
    payload = {
        "model": "qwen2.5-7b-instruct",
        "messages": [
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": f"Вход: {text}"}
        ],
        "temperature": 0.0 # Ноль делает модель роботом: она не фантазирует, а строго исполняет правила
    }
    
    try:
        response = requests.post(url, json=payload)
        result_text = response.json()['choices'][0]['message']['content']
        
        # Вытаскиваем JSON из ответа (даже если нейросеть добавит лишние символы)
        clean_json = re.search(r'\{.*\}', result_text, re.DOTALL)
        if clean_json:
            parsed = json.loads(clean_json.group(0))
        else:
            parsed = json.loads(result_text)
            
        return parsed.get("Продукт единый", "Ошибка формата")
            
    except Exception as e:
        print(f"Ошибка парсинга: {e} | Ответ нейросети был: {result_text}")
        return "Ошибка"

# 4. Запускаем обработку
print("Запускаю анализ (ИИ читает строки)...")

# Применяем функцию и создаем колонку "Продукт единый"
df_test['Продукт единый'] = df_test[column_to_analyze].apply(ask_qwen)

# Оставляем ТОЛЬКО 2 нужные колонки
df_final = df_test[[column_to_analyze, 'Продукт единый']]

# Выводим результат на экран
print("Готово! Результат работы нейросети:")
display(df_final)

# Если захотите сохранить результат в новый Excel файл, раскомментируйте строчку ниже:
# df_final.to_excel(r'O:\chemistry\data_analyzed.xlsx', index=False)
```
